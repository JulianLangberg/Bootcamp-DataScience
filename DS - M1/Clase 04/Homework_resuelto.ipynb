<div><b style="color:#ff7d00;">A partir del CSV hospitales2.csv, generar un archivo CSV de salida, que contenga los siguientes campos en este orden:</br></br>

<div style="color:#ff006d;">latitude</br>
longitude</br>
name</br>
label</br></br>

<div style="color:#ff7d00;">Con los correspondientes datos extraídos del CSV original, el campo name tiene que corresponder con la dirección del hospital, y el campo label con el nombre del hospital.</div>

```python
import pandas as pd
```

```python
hospital=pd.read_csv('hospitales2.csv')
hospital.head(5)
```

```python
hospital['WKT']=hospital['WKT'].str.replace('(','')
hospital['WKT']=hospital['WKT'].str.replace(')','')
hospital['WKT']=[i.split(sep=' ')[1:] for i in hospital['WKT']]
hospital.head(5)
```

```python
hospital['latitude']=[hospital['WKT'][i][0] for i in hospital.index]
hospital['longitude']=[hospital['WKT'][i][1] for i in hospital.index]
```

```python
hospital.head(5)
```

```python
hospital.head(10)
```

```python
columnas=hospital.columns
conservar=['DOM_NORMA','NOMBRE','latitude','longitude']
columnas=[i for i in columnas if not(i in conservar)]
columnas
```

```python
hospital=hospital.drop(columnas, axis=1)
```

```python
hospital.reindex(columns=['latitude','longitude','DOM_NORMA','NOMBRE'])
```

```python
hospital.rename(columns={'latitude':'latitude','longitude':'longitude','DOM_NORMA':'name','NOMBRE':'label'})
```

```python
hospital.to_csv(path_or_buf='Solución_hospital')
```
